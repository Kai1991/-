### 数据与特征处理

* 数据采集 ： 这里不再赘述

* 数据清洗 ： 
    * 从业务角度思考，一些异常的业务数据
    * 组合或同级属性判定 ： 比如一个人要买篮球鞋，但是样本中女性用户有85%
    * 缺省值：有些样本的缺省值极多，需要删除相关字段
    
* 数据采样 ： 
    * 很多情况下，样本数据不均衡，到那会有些模型对样本均衡很敏感（比如：LR）
    * 采样的手段有：随机采样，分层采样，上采样，下采样和损失函数修改
    
* 特征处理 ：
    * 数值型：
        * 幅度调整，归一化（最大最小归一化,标准化）
        * log变化
        * 统计值处理（max,min,mean,std）
        * 离散化
        * hash分桶
        * 每个类别下对应的变量统计值
    
    * 类别型：
        * one-hot编码/哑变量
        * hash和聚类处理
        * 统计每个类别变量下各个target的比例，转成数值型
    
    * 时间型：
        * 特点：既可以看成连续值也可以看成离散值
        * 连续性：
            * 持续了多长时间
            * 间隔时间
        * 离散性：
            * 一天中哪个时间段
            * 一周中的星期几
            * 一年中哪个星期
            * 一年中哪个季度
            * 工作日，周末
        * 最近第1/2/3/7的行为统计与品骏行为统计
    
    * 文本型：
        * 词袋
        * n-gram
        * TF-IDF: TF: 词在当前文中出现的次数 / 在全部文档中出现的次数。IDF:ln(总文档数/含词的文档书）
        * word2vec
    
    * 统计型：(重点)
        * 加减平均：商品价格高于平均价格多少，用户在摸个商品类目下消费超过平均用户多少，用户理想奶登录天数超过平均多少
        * 分位数：商品属于输出商品价格的多少分位线处
        * 次序型：排在第几位
        * 比例类
        * 商户维度统计 mean,max,min,std
        * 商品维度统计 mean,max,min,std
    
    * 组合特征：
        * 特征简单拼接
        * 特征组合：
            * 是用GBDT产生特征组合路径
            * 组合特征和原特征范进LR训练

### 特征选特征
需要进行特征选择的原因： 部分特征相关度太高，消耗计算性能，部分特征对预测结果有付影响

* 过滤型：计算特征与结果值的相关度，排序留下topN相关的特征，缺点：没有考虑到 特征之间的组合关系

* 包裹型： 使用特征子集，用模型判断效果，进行迭代，如果下过突然变差，迭代停止

* 嵌入型： 使用树型模型分析特征的重要程度
