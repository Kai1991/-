### 1.概率论基础

* 事件：一个随机的过程的结果
    * 例如，掷骰子，可能的结果有{1,2,3,4,5,6}

* 独立事件：A和B的发生不相互影响
    * $P(A \bigcap B) = P(A) * P(B)$
    
* 条件概率：P(Y|X) 
    * 直观上，在X发生的情况下，发生Y的概率

* 联合概率 P(XY) = P(X)* P(Y|X)
    * X和Y 同事发生的概率
    * $ P(X_1 ... X_n) = P(X_1)P(X_2|X_1)P(X_3|X_2X_1) ... P(X_n|X_{n-1}X_{n_2}...X_1) $
    * 马尔科夫 ： $ P(X_1,...,X_n) = P(X_1)P(X_2|X_1)P(X_3|X_2)...P(X_n|X_{n-1}) $
    
* 贝叶斯公式： P(XY) = P(X|Y)* P(Y) = P(Y|X)* P(X)
    * 或者这种表达形式：$ P(Y|X) = { {P(X|Y)P(Y)} \over P(X) }$
    * P(Y|X) 后验概率
    * P(Y) 先验概率
  
* 随机变量分布
    * 伯努利分布 ： (二分类) P(Y=1) = P = 1-P(Y=0)=1-q
    * 分类分布：（多分类），多个离散值，参数是（P1,...,Pn）
    
    * 二项分布：n次伯努利，k次成功的概率
    * 多项分布：n次分类分布，(K1,K2,,,,Kn)的概率
    
* 期望：假设，x1,x2,x3..xn对应的概率是p1,...,pn 那么X的期望为 E[X] = x1px + x2p2,...xnpn

* 方差：$ Var[X] = E[(X - \mu )] $ 可以理解成 离平均数的平均偏离值

#### 连续随机变量

* 概率 $ P(X) = \int_s f(x)dx $

* 期望：$ E[X] = \int Xf(X)dx $

* 方差：$ Var[X] = \int (X - \mu )^2f(x)dx $

* 协方差和相关系数
    * 协方差 COV（X,Y）= E[ ( X - E[X])(Y - E[Y])] = E[XY] -E[X]E[Y]
    * 相关系数 $ COR(X,Y) = COV(X,Y) \over \sqrt{Var(X) Var(Y)} $


* 熵 ： = $ - \sum P(X_i)logP(X_i) $
