### 1.基本理论

#### 1.1线性代数中的基本概念
矩阵是线性代数提供的一种紧凑表示和操作线性方程组的方法

#### 1.2矩阵计算及性质

* 向量计算
    * 向量向量乘积-內积：行向量乘列向量得到一个常数。表示形式：$ \lt A,B\gt = [a_1,a_2,...,a_n][ b_1,b_2,...,b_n]^T = \sum_{j=1}^n a_jb_j $ 
    * 向量向量乘积-外积：列向量乘行向量得到一个矩阵。表示形式 $ \lt A,B\gt = [a_1,a_2,...,a_n]^T[ b_1,b_2,...,b_n] = 矩阵 $
    
    * 向量的內积与正交性：$ \lt A,B\gt = ||A||||B||cos(\theta) $ 如果$ \lt A,B\gt = 0 $ 则A与B垂直 

* 矩阵计算

    * 矩阵加法：对应位置元素相加

    * 矩阵数乘：给定常数，这个成熟乘矩阵等于 矩阵中的每个元素乘以这个常数
    
    * 矩阵乘法：矩阵A,B 矩阵A的行乘以矩阵B的列之后组合成新的矩阵

    * 矩阵的正交阵：如果方阵A 有 $A^T A= I 或者 A^=A^{-1}$ 则A方阵为正交矩阵
    
* 矩阵-向量计算
    * 给定矩阵A(m行n列) 和向量x(n行) Ax 可以看做是矩阵A各列的线性组合
    
* 矩阵的转置
    * 转置：沿主对角线交换数据
    
* 特殊矩阵： 上矩阵，下矩阵，对角阵，单位阵，对称阵

* 矩阵的迹： 矩阵对角线元素的和

* 矩阵的列向量线性无关性：矩阵A,向量x,如果 Ax = 0 则表示 矩阵 向量组线性无关

* 矩阵的秩：

    * 矩阵线性无关的列向量，；列向量空间的任何一个向量都可用这些线性无关的向量表示，则线性无关的列向量的个为向量的秩 ：rank(A)
    
    * 如果矩阵A(n*n)为方阵，切矩阵的秩rank(A) = n 则矩阵A为非奇异
    
    * 如果方阵A可逆，则方阵A 非奇异
    
* 方阵特性：
    * 特征值，特征向量： 如果n阶方阵A,数$\lambda$ 和n维向量x,满足:
        * $Ax = \lambda x$
    * A矩阵的特征值：$\lambda$ ,特征向量：x
